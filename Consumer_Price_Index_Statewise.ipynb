
# State Level Consumer price index Rural Urban upto May 2023

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.simplefilter(action='ignore',category=FutureWarning)
```

```python
DATA=pd.read_csv("State Level Consumer price index RuralUrban upto May 2023.csv")
```

```python
DATA
```

```python
DATA.columns
```

```python
DATA=DATA.drop(['Arunachal Pradesh', 'Bihar', 'Delhi','Nagaland','Rajasthan', 'Sikkim','Telangana','Uttarakhand'],axis=1)
```

```python
DATA.columns
```

```python
DATA=DATA.dropna()
```

```python
pd.isnull(DATA).sum()
```

```python

sns.barplot(x='Year',y='Assam',data=DATA,hue='Sector')
plt.show()
```

```python
sns.lineplot(x='Year',y='Andhra Pradesh',data=DATA,hue='Sector')
plt.figure(figsize=(10,5))
plt.show()
```

```python
fg=sns.FacetGrid(DATA,col='Sector',hue='Year')
fg.map(plt.plot,"Chattisgarh").add_legend()
plt.show()
```

```python
# sns.set(style="darkgrid")
sns.boxplot(x="Year",y="Goa",data=DATA,hue='Sector',palette="twilight_r")
```

```python
sns.set(style="whitegrid")
sns.stripplot(data=DATA, x="Year",y="Gujarat",hue="Sector",palette="autumn",linewidth=1,size=8)
# plt.grid()
```

```python
sns.catplot(x="Year",y="Haryana",data=DATA,hue='Sector',kind="boxen",palette="coolwarm",height=6)
```

```python
Hp=DATA[['Sector','Year','Himachal Pradesh','Jharkhand','Karnataka']]
Hp
```

```python
sns.pairplot(data=Hp,hue="Sector", palette="Dark2", kind='scatter',diag_kind="kde")
```

# Dashboard 

```python
f,axes= plt.subplots(2,2, figsize=(15,15))
sns.set_style("whitegrid")
plt.title("Consumer_Price_Index")

fig1=sns.lineplot(x='Year',y='Andhra Pradesh',data=DATA,hue='Sector',ax=axes[0,0])

sns.set(style="whitegrid")
fig2=sns.stripplot(data=DATA, x="Year",y="Gujarat",hue="Sector",palette="autumn",linewidth=1,size=8,ax=axes[0,1])

fig3=sns.barplot(x='Year',y='Assam',data=DATA,hue='Sector',ax=axes[1,1])

fig4=sns.boxplot(x="Year",y="Goa",data=DATA,hue='Sector',palette="twilight_r",ax=axes[1,0])


fg=sns.FacetGrid(DATA,col='Sector',hue='Year')
fg.map(plt.plot,"Chattisgarh").add_legend()

sns.pairplot(data=Hp,hue="Sector", palette="Dark2", kind='scatter',diag_kind="kde")


plt.tight_layout()
sns.despine()
plt.show()
```
